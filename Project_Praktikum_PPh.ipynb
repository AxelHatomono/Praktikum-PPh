## Data wrangling

### Subtask:
Data Wrangling: Feature Engineering and Data Transformation


**Reasoning**:
Perform feature engineering, data type conversion, calculate profit, and handle outliers as requested in the subtask.



```python
# Feature Engineering from Dates
df = pd.get_dummies(df, columns=['tahun'], prefix='year')

# Data Type Conversion
# Check and convert numerical columns with potential formatting issues
for col in ['pendapatan', 'beban_operasional', 'penyusutan', 'taxable_income', 'tax_liability', 'total_assets', 'total_nilai_perolehan',
            'total_metode_garis_lurus', 'total_metode_saldo_menurun', 'total_depreciation_difference']:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric, invalid parsing will be NaN
    except:
        print(f"Error converting column {col} to numeric. Check data format.")


# Create profit column
df['profit'] = df['pendapatan'] - df['beban_operasional']


# Outlier handling in 'profit'
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['profit'])
plt.title('Boxplot of Profit')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['profit'], kde=True)
plt.title('Distribution of Profit')
plt.show()

# Calculate IQR
Q1 = df['profit'].quantile(0.25)
Q3 = df['profit'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['profit'] < lower_bound) | (df['profit'] > upper_bound)]
print("Outliers:\n", outliers)

# Decide on outlier handling (e.g., clipping)
# Clipping outliers to the bounds
df['profit'] = df['profit'].clip(lower=lower_bound, upper=upper_bound)

# Visualize post-handling
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['profit'])
plt.title('Boxplot of Profit (after clipping)')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['profit'], kde=True)
plt.title('Distribution of Profit (after clipping)')
plt.show()
```
